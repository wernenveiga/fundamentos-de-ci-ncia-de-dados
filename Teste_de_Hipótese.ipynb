#Teste de Hipótese

##Bibliotecas

```python
import pandas as pd
import numpy as np
from scipy import stats
#Hipotese T
from scipy.stats import t
from scipy.stats import ttest_ind
#Hipotese Chi Quadrado
import scipy.stats
from scipy.stats import chi2
from scipy.stats import chi2_contingency
```

```python
#Arquivo

arquivo = "/content/mamalia_sleep.csv"

df_mamiferos = pd.read_csv(arquivo)
```

##Teste de Hipótese Z

```python
med_sono_pop = df_mamiferos['sleep_total'].mean()
despd_sono_pop = df_mamiferos['sleep_total'].std()
df_mamiferos['sleep_total'].describe()
```

```python
sono_ordem = df_mamiferos.groupby('order')['sleep_total'].mean() #Media de sonos por ordem
sono_ordem
```

```python
is_rodentia = df_mamiferos['order'] == "Rodentia"
df_rodentia = df_mamiferos[is_rodentia]
```

```python
sono_roedores = df_rodentia['sleep_total']
med_sono_roedores = sono_roedores.mean()
```

```python
##Calcular o Z
z_value = (med_sono_roedores - med_sono_pop)/ despd_sono_pop
z_value
```

```python
stats.norm.cdf(z_value)
(1 - stats.norm.cdf(z_value))*100
```

## Teste de Hipótese T

```python
#Arquivo

arquivo = "/content/salarios.csv"

df_salarios = pd.read_csv(arquivo)
df_salarios
```

```python
df_salarios['INCOME'].describe()
```

```python
is_white = df_salarios['RACE'] == "White"
is_white
```

```python
df_brancos = df_salarios[is_white]
df_brancos
```

```python
salario_brancos = df_brancos['INCOME']
salario_brancos
```

```python
is_not_white = df_salarios['RACE'] != 'White'
is_not_white
```

```python
df_not_white = df_salarios[is_not_white]
df_not_white
```

```python
salario_nao_brancos = df_not_white['INCOME']
salario_nao_brancos
```

```python
salario_brancos.describe()
```

```python
salario_nao_brancos.describe()
```

```python
t_stat, p_value = ttest_ind(salario_brancos, salario_nao_brancos)
print(t_stat, p_value)
print("t = %.3f, p = %.6f" % (t_stat, p_value))
```

```python
count_salariosbrancos, count_salariosnaobrancos = len(salario_brancos), len(salario_nao_brancos)
grau_liberdade = count_salariosbrancos + count_salariosnaobrancos - 2
grau_liberdade
```

```python
alpha = 0.05 #95% de confiabilidade
valor_critico = t.ppf(1.0 - alpha, grau_liberdade)
valor_critico
```

```python
if abs(t_stat) <= valor_critico: #Teste usando o valor de T
  print("Aceita-se a Hipótese Nula. Não há diferença significativa de salários")
else:
  print("Rejeita-se a Hipótese Nula. Há diferença significativa de salários")
```

```python
if p_value > alpha: #Teste usando o valor de P sendo maior se comparando com o alpha
  print("Aceita-se a Hipótese Nula. Não há diferença significativa de salários")
else:
  print("Rejeita-se a Hipótese Nula. Há diferença significativa de salários")
```

## Tesde Hipótese Chi Quadrado

```python
#Arquivo

arquivo = '/content/titanic.csv'

df_titanic = pd.read_csv(arquivo)
df_titanic
```

```python
df_titanic.columns
```

```python
df_sobreviventes_sex = df_titanic[['Survived', 'Sex']] #pegando só a coluna sobrevivemntes por sexo
df_sobreviventes_sex
```

```python
scipy.stats.chisquare(df_sobreviventes_sex['Sex'].value_counts())
```

```python
scipy.stats.chisquare(df_sobreviventes_sex['Survived'].value_counts())
```

```python
freq_table = pd.crosstab(df_sobreviventes_sex['Sex'], df_sobreviventes_sex['Survived'])
freq_table
```

```python
chi2stat, p_value, dof, freq_experada = chi2_contingency(freq_table)
print(chi2stat)
print(p_value)
print(dof)
print(freq_experada)
```

```python
prob = 0.95
valor_crit = chi2.ppf(prob, dof)
valor_crit
```

```python
if abs(chi2stat) >= valor_crit:
  print("Sexo influênciou na sobrevivência (H0 rejeitada)")
else:
  print("Sexo não influênciou na sobrevivênia (H0 Aceita)")
```

```python
alpha = 1.0 - prob
if p_value <= alpha:
  print("Sexo influênciou na sobrevivência (H0 rejeitada)")
else:
  print("Sexo não influênciou na sobrevivênia (H0 Aceita)")
```
